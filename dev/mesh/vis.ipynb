```python
import xarray as xr
import matplotlib.pyplot as plt
# Load the NetCDF file
file_path = '.data/MESH_initial_values.nc'

with xr.open_dataset(file_path) as ds:
    # Display the dataset information
    ds["grid_qi"] = ds["grid_qi"] * 4
    ds["grid_qo"] = ds["grid_qi"] * 4

    ds.to_netcdf(".data/MESH_modified_values.nc", mode='w', format='NETCDF4')
    
```

```python
with xr.open_dataset('.data/MESH_modified_values.nc') as ds:
    print(ds)
    ds['grid_qi'].plot()
    plt.title("Modified grid_qi")
    plt.show()

    ds['grid_qo'].plot()
    plt.title("Modified grid_qo")
    plt.show()
```
