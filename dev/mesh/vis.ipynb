```python
import xarray as xr

# Load the NetCDF file
file_path = '.data/MESH_initial_values.nc'
dataset = xr.open_dataset(file_path)

# Display the dataset
print(dataset)

import matplotlib.pyplot as plt

# Plot a variable from the dataset (e.g., 'grid_qi')
variable_to_plot = 'grid_qo'

# Ensure the variable exists in the dataset
if variable_to_plot in dataset:
    (5*dataset[variable_to_plot]).plot()
    plt.title(f"Plot of {variable_to_plot}")
    plt.show()
else:
    print(f"Variable '{variable_to_plot}' not found in the dataset.")
```

```python
dataset['grid_qi'] = dataset['grid_qi'] * 4
dataset['grid_qo'] = dataset['grid_qo'] * 4

dataset.to_netcdf('.data/MESH_modified_values.nc')
```
