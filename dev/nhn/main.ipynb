```python
from ftplib import FTP, error_perm
from pathlib import Path

# Cell to download all .gml files from the specified FTP directory

HOST = "ftp.maps.canada.ca"
REMOTE_DIR = "pub/nrcan_rncan/vector/geobase_nhn_rhn/gml_en/09"
OUT_DIR = Path("gml_files")
OUT_DIR.mkdir(parents=True, exist_ok=True)

ftp = FTP(HOST, timeout=60)
ftp.login()  # anonymous
try:
    ftp.cwd(REMOTE_DIR)
except error_perm as e:
    ftp.quit()
    raise RuntimeError(f"Could not change directory to {REMOTE_DIR}: {e}")

try:
    names = ftp.nlst()
except error_perm as e:
    # Some FTP servers return error for empty dirs; handle gracefully
    names = []
    print("Failed to list directory or directory empty:", e)

gml_files = [n for n in names if n.lower().endswith(".zip")]

if not gml_files:
    print("No .gml files found in", REMOTE_DIR)
else:
    for fname in gml_files:
        dest = OUT_DIR / fname
        if dest.exists():
            print("Skipping (exists):", fname)
            continue
        print("Downloading:", fname)
        with open(dest, "wb") as f:
            try:
                ftp.retrbinary(f"RETR {fname}", f.write)
            except Exception as e:
                print("Failed to download", fname, ":", e)
                if dest.exists():
                    dest.unlink()

ftp.quit()
print("Done. Files saved in:", OUT_DIR.resolve())
```

```python
import geopandas as gpd
import zipfile
import random

import matplotlib.pyplot as plt

# OUT_DIR and gml_files are defined in a previous cell
zip_paths = [OUT_DIR / name for name in gml_files]

gdfs = []
for zp in zip_paths:
    if not zp.exists():
        print("Zip not found, skipping:", zp)
        continue
    try:
        with zipfile.ZipFile(zp, "r") as z:
            # find gml files inside the zip
            gml_names = [n for n in z.namelist() if n.lower().endswith(".gml")]
            if not gml_names:
                print("No .gml inside:", zp.name)
                continue

            for gml_name in gml_names:
                # Use GDAL/OGR virtual filesystem for zipped contents
                vsizip_path = f"/vsizip/{zp.resolve().as_posix()}/{gml_name}"
                try:
                    gdf = gpd.read_file(vsizip_path)
                    # add origin info for plotting/legend
                    gdf["_source_zip"] = zp.name
                    gdf["_source_gml"] = gml_name
                    gdfs.append(gdf)
                    print("Loaded:", zp.name, "->", gml_name, " (rows:", len(gdf), ")")
                except Exception as e:
                    print("Failed to read", vsizip_path, ":", e)
    except Exception as e:
        print("Failed to open zip", zp, ":", e)

if not gdfs:
    print("No GML layers loaded.")
else:
    # unify CRS: use the first loaded gdf's CRS as target (if available)
    target_crs = gdfs[0].crs
    if target_crs is not None:
        for i in range(len(gdfs)):
            if gdfs[i].crs != target_crs:
                try:
                    gdfs[i] = gdfs[i].to_crs(target_crs)
                except Exception as e:
                    print("CRS transform failed for", gdfs[i].get("_source_zip", "?"), ":", e)

    # plot all layers on one axis with different colors
    fig, ax = plt.subplots(figsize=(12, 10))
    for gdf in gdfs:
        label = f"{gdf['_source_zip'].iloc[0]}:{gdf['_source_gml'].iloc[0]}"
        color = (random.random(), random.random(), random.random())
        # if geometry type is mixed, plot without legend entry for cleanliness
        try:
            gdf.plot(ax=ax, color=color, alpha=0.6, edgecolor="k", linewidth=0.2, label=label)
        except Exception:
            # fallback: plot only geometry column (some gml layers may contain non-geometry columns interfering)
            gpd.GeoSeries(gdf.geometry).plot(ax=ax, color=color, alpha=0.6, edgecolor="k", linewidth=0.2)

    ax.set_title("GML layers from zipped files in: " + str(OUT_DIR))
    ax.legend(loc="best", fontsize="small")
    ax.set_axis_off()
    plt.show()
```
